```python
def hello():
    print('hello python')
hello()
```

```python
def add(a,b):
    print(a+b)
add(4,5)
```

```python
def count_vowel():
    s=input('enter a string')
    vowel='aeiouAEIOU'
    count=0
    for ch in s:
        if ch in vowel:
            count+=1
    print(count)
count_vowel()
```

```python
def is_prime(n):
    if n<=1:
        print('not prime')
        return
    i=2
    while i<n:
        if n%2==0:
            print('not a prime')
            return
        i+=1
    print('prime')
is_prime(8)
```

```python
def factorial(n):
    fact=1
    for i in range(1,n+1):
        fact*=i
    print("factorial=",fact)
factorial(4)
```

```python
def reverse_string(s):
    rev=''
    for ch in s:
        rev=ch+rev
    print(rev)
reverse_string('python')
```

```python
def palindrom(s):
    rev=s[::-1]
    if s==rev:
        print('palindrom')
    else:
        print('not a palindrom')
palindrom('level')
```

```python
def fact(n):
    if n==0 or n==1:
        return 1
    return n*fact(n-1)
print(fact(5))
```

```python
def sum(n):
    if n==0:
        return 0
    return n+sum(n-1)
sum(5)
```

```python
def fib(n):
    if n<=1:
        return n
    return fib(n-1)+fib(n-2)
fib(6)
```

```python
def check(n):
    n=int(input('enter number='))
    if n%2==0:
        return 'even'
    else:
        return 'odd'
check(3)
```

```python
def square(n):
    return n**3
square(5)
```

```python
def fact(n):
    f=1
    i=1
    while i<=n:
        f*=i
        i+=1
    return f
fact(5)
```

```python
def greatest(a,b,c):
    if a>b and a>c:
        return a
    elif b>c:
        return b
    else:
        return c
greatest(2,3,4)
```

```python
class student:
    def __init__(self,name,marks):
        self.name=name
        self.marks=marks
    def welcome(self):
        print('welcome',self.name)
    def get_marks(self):
        return self.marks
s1=student('bheem',90)
s1.welcome()
print('marks=',s1.get_marks())
```

```python
class student:
    def __init__(self,name,marks):
        self.name=name
        self.marks=marks
    def get_avg(self):
        sum=0
        for val in self.marks:
            sum+=val
        print('hii',self.name,'your avg marks is=',sum/3)
s1=student('bheem',[98,90,89])
s1.get_avg()
```

```python
class car:
    colour='black'
    @staticmethod
    def start():
        print('car started...')
    @staticmethod
    def stop():
        print('car stopped...')
class toyotacar(car):
    def __init__(self,name):
        self.name=name
c1=toyotacar('fortuner')
print(c1.name)
c1.start()
c1.stop()
```
