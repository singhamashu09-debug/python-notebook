```python
import pandas as pd
import matplotlib.pyplot as plt
data={
    "patient_name":["amit","riya","suresh","neha","rahul","pooja"],
    "Age":[25,34,45,29,52,41],
    "Gender":["male","female","male","female","male","female"],
    "Disease":["fever","diabetes","bp","cold","heart","asthma"],
    "hospital visit":[2,5,3,4,6,5]
}
df=pd.DataFrame(data)
df.to_csv("healthcare_dataset.csv", index=False)
df=pd.read_csv("healthcare_dataset.csv")
```

```python
import matplotlib.pyplot as plt

patient_names = ["amit","riya","suresh","neha","rahul","pooja"]
hospital_visits = [2,5,3,4,6,5]

plt.scatter(patient_names, hospital_visits, color='darkorange', s=100, edgecolors='black')
plt.xlabel("Patient Name")
plt.ylabel("Hospital Visits")
plt.title("Hospital Visits per Patient")
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```python
plt.hist(df["Age"], bins=5, color="orange", edgecolor="black")
plt.xlabel("Age")
plt.ylabel("Count")
plt.title("Age Distribution of Patients")
plt.show()
```

```python
bars=plt.bar(df["patient_name"], df["hospital visit"], color="blue")
plt.xlabel("Patient Name")
plt.ylabel("Hospital Visits")
plt.title("Hospital Visits per Patient")
patterns=['/','o','*','.','-','+']
for bar in bars:
    bar.set_hatch(patterns.pop(0))
# plt.legend()
plt.show()
```

```python
df.groupby("Gender")["hospital visit"].sum().plot(kind="bar", color=["blue","pink"])
plt.xlabel("Gender")
plt.ylabel("Total Hospital Visits")
plt.title("Hospital Visits by Gender")
plt.show()
```

```python
import matplotlib.pyplot as plt

patient_name = ["amit","riya","suresh","neha","rahul","pooja"]
hospital_visit =  [2,5,3,4,6,5]

# Use indices for plotting
plt.plot(range(len(patient_name)), hospital_visit, 'b.--', label='Visits')

plt.xticks(range(len(patient_name)), patient_name)  # Replace indices with names
plt.xlabel("Patient Name")
plt.ylabel("Number of Hospital Visits")
plt.title("Hospital Visits per Patient")
plt.legend()
plt.show()
```
